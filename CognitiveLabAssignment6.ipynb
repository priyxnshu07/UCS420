Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
 Generate x values from -10 to 10 using np.linspace().
 Compute y values for:
o y= M⋅x
2
o y=M⋅sin(x)
 Plot both functions in a single figure:
o Use different colors and line styles.
o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt

# User input for M
M = float(input("Enter a value for M: "))

# Generate x values
x = np.linspace(-10, 10, 400)

# Compute y values
y1 = M * x**2
y2 = M * np.sin(x)

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label=r"$y=Mx^2$", color='r', linestyle='--')
plt.plot(x, y2, label=r"$y=M\sin(x)$", color='b', linestyle='-')

# Customization
plt.title("Mathematical Function Plots")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid()
plt.show()
```

Q.2 Create a dataset of five subjects and scores.
 Convert it into a Pandas DataFrame.
 Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.


```python
import pandas as pd
import seaborn as sns

# Create dataset
data = {'Subjects': ['Math', 'Physics', 'Chemistry', 'Biology', 'CS'],
        'Scores': [85, 78, 92, 74, 88]}

df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='Subjects', y='Scores', data=df, palette="viridis")

# Add annotations
for index, row in df.iterrows():
    plt.text(index, row['Scores'] + 1, str(row['Scores']), ha='center', fontsize=12)

# Customization
plt.title("Student Scores in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis='y')
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
 Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
 Customize with titles, axis labels, and grids.


```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed (replace with your roll number if needed)
roll_number = 12345  # Change this to your actual roll number
np.random.seed(roll_number)

# Generate random dataset
data = np.random.randn(50)

# Create subplots
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# Line plot (Cumulative Sum)
ax[0, 0].plot(np.cumsum(data), color='b', marker='o', linestyle='-')
ax[0, 0].set_title("Cumulative Sum")
ax[0, 0].set_xlabel("Index")
ax[0, 0].set_ylabel("Cumulative Value")
ax[0, 0].grid()

# Scatter plot with random noise
ax[0, 1].scatter(range(50), data, color='r', marker='x')
ax[0, 1].set_title("Scatter Plot with Random Noise")
ax[0, 1].set_xlabel("Index")
ax[0, 1].set_ylabel("Random Values")
ax[0, 1].grid()

plt.tight_layout()
plt.show()
```

Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Download dataset
url = "https://github.com/AnjulaMehto/MCA/raw/main/company_sales_data.csv"
df = pd.read_csv(url)

# 1. Line plot for Total Profit
plt.figure(figsize=(8, 5))
sns.lineplot(x=df["month_number"], y=df["total_profit"], marker="o", color="b")
plt.title("Total Profit per Month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.grid()
plt.show()

# 2. Multi-line plot for product sales
plt.figure(figsize=(8, 5))
for column in df.columns[1:-1]:  # Exclude month_number & total_profit
    sns.lineplot(x=df["month_number"], y=df[column], label=column)

plt.title("Product Sales per Month")
plt.xlabel("Month Number")
plt.ylabel("Sales")
plt.legend()
plt.grid()
plt.show()

# 3. Bar chart for all attributes
df.set_index("month_number").plot(kind="bar", figsize=(10, 6), colormap="viridis")
plt.title("Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Values")
plt.grid(axis='y')
plt.show()
```
