# Lab Assignment 7: E-Commerce Sales Data Analysis
## Objective:
This notebook analyzes randomized sales data for an e-commerce company, performs statistical analysis, and generates visual insights to help with business decision-making.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Part I: Randomized Sales Data Generation

```python
# Step 1: Initialize Random Seed (Using the roll number)
roll_number = 102497011  # Unique seed for reproducibility
np.random.seed(roll_number)

# Step 2: Generate Sales Data (Random values between 1000 and 5000)
sales_data = np.random.randint(1000, 5000, size=(12, 4))

# Step 3: Convert to Pandas DataFrame
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df = pd.DataFrame(sales_data, columns=categories, index=months)

# Display first 5 rows
display(df.head())
```

## Part II: Data Manipulation and Analysis

```python
# Step 1: Calculate total sales per category
total_sales_category = df.sum()

# Step 2: Calculate total sales per month
df['Total Sales'] = df.sum(axis=1)

# Step 3: Calculate average sales growth between consecutive months
df['Growth Rate'] = df['Total Sales'].pct_change() * 100  # Percentage growth

# Step 4: Apply Discount based on Roll Number
if roll_number % 2 == 0:
    df['Electronics'] *= 0.9
else:
    df['Clothing'] *= 0.85
display(df)
```

## Part III: Visualizations

```python
# Step 1: Line Plots for Monthly Sales Trends
plt.figure(figsize=(10, 5))
for category in categories:
    sns.lineplot(x=months, y=df[category], label=category)
plt.title("Monthly Sales Trends by Category")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid()
plt.show()
```

```python
# Step 2: Box Plots for Sales Distribution
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution by Category")
plt.xlabel("Product Categories")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percenle of flaened array, for each column, for each
row.
iii. Mean, Median and Standard Deviation of flatened array, of each column, and
each row

```python
import numpy as np

# Define the array
array = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Element-wise absolute value
abs_array = np.abs(array)
print("Element-wise absolute value:\n", abs_array)

# ii. Percentiles (25th, 50th, 75th)
percentiles_flat = np.percentile(array, [25, 50, 75])  # For flattened array
percentiles_columns = np.percentile(array, [25, 50, 75], axis=0)  # For each column
percentiles_rows = np.percentile(array, [25, 50, 75], axis=1)  # For each row

print("\nPercentiles for flattened array:\n", percentiles_flat)
print("\nPercentiles for each column:\n", percentiles_columns)
print("\nPercentiles for each row:\n", percentiles_rows)

# iii. Mean, Median, and Standard Deviation
mean_flat = np.mean(array)
median_flat = np.median(array)
std_flat = np.std(array)

mean_columns = np.mean(array, axis=0)
median_columns = np.median(array, axis=0)
std_columns = np.std(array, axis=0)

mean_rows = np.mean(array, axis=1)
median_rows = np.median(array, axis=1)
std_rows = np.std(array, axis=1)

print("\nMean, Median, and Standard Deviation for flattened array:")
print("Mean:", mean_flat, "Median:", median_flat, "Standard Deviation:", std_flat)

print("\nMean, Median, and Standard Deviation for each column:")
print("Mean:", mean_columns, "\nMedian:", median_columns, "\nStandard Deviation:", std_columns)

print("\nMean, Median, and Standard Deviation for each row:")
print("Mean:", mean_rows, "\nMedian:", median_rows, "\nStandard Deviation:", std_rows)
```

Q.3 For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).
i. Find floor, ceiling and truncated value, rounded values

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)
ceiling_values = np.ceil(a)
truncated_values = np.trunc(a)
rounded_values = np.round(a)

print("Floor values:", floor_values)
print("Ceiling values:", ceiling_values)
print("Truncated values:", truncated_values)
print("Rounded values:", rounded_values)
```

Q.4 Swap Two Elements in a List Using a Temporary Variable
python



```python
my_list = [10, 20, 30, 40, 50]

index1, index2 = 1, 3  # Swap elements at index 1 and 3

temp = my_list[index1]
my_list[index1] = my_list[index2]
my_list[index2] = temp

print("List after swapping:", my_list)
```

Q.5 Swap Two Elements in a Set by Converting it to a List

```python
my_set = {10, 20, 30, 40, 50}

list_version = list(my_set)  # Convert to list
index1, index2 = 1, 3  # Choose indices to swap

list_version[index1], list_version[index2] = list_version[index2], list_version[index1]

swapped_set = set(list_version)  # Convert back to set
print("Set after swapping:", swapped_set)
```
