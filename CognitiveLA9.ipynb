```python
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer
nltk.download('wordnet')

ps = PorterStemmer()
ls = LancasterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed_porter = [ps.stem(word) for word in filtered_words]
stemmed_lancaster = [ls.stem(word) for word in filtered_words]
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("Porter Stemmer:\n", stemmed_porter)
print("\nLancaster Stemmer:\n", stemmed_lancaster)
print("\nLemmatized Words:\n", lemmatized_words)
```

```python
import re

long_words = re.findall(r'\b[a-zA-Z]{6,}\b', text)
numbers = re.findall(r'\d+', text)
capitalized_words = re.findall(r'\b[A-Z][a-z]*\b', text)
alpha_words = re.findall(r'\b[a-zA-Z]+\b', text)
vowel_words = re.findall(r'\b[aeiouAEIOU][a-zA-Z]*\b', text)

print("Words with >5 letters:\n", long_words)
print("\nExtracted Numbers:\n", numbers)
print("\nCapitalized Words:\n", capitalized_words)
print("\nAlphabetic Words Only:\n", alpha_words)
print("\nWords Starting with Vowels:\n", vowel_words)
```

```python
def custom_tokenizer(text):
    text = re.sub(r'[^\w\s\.\'-]', '', text)
    return re.findall(r"\b[\w'-]+\b", text)

custom_tokens = custom_tokenizer(text)

text_with_entities = """
Contact us at support@example.com or visit https://www.example.com.
Call +91 9876543210 or 123-456-7890 for assistance.
"""

text_cleaned = re.sub(r'\b[\w.-]+@[\w.-]+\.\w+\b', '<EMAIL>', text_with_entities)
text_cleaned = re.sub(r'https?://\S+', '<URL>', text_cleaned)
text_cleaned = re.sub(r'\+?\d{1,2}[-.\s]?\d{10}|\d{3}[-.\s]?\d{3}[-.\s]?\d{4}', '<PHONE>', text_cleaned)

print("Custom Tokens:\n", custom_tokens)
print("\nText After Substitutions:\n", text_cleaned)
```
